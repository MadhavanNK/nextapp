```python
df.drop(['DailyRate',
         'RelationshipSatisfaction',
         'YearsSinceLastPromotion',
         'Education',
         'PercentSalaryHike',
         'WorkLifeBalance',
         'TrainingTimesLastYear',
         'NumCompaniesWorked',
         'HourlyRate', 
         'EmployeeCount',
         'EmployeeNumber', 
         'MonthlyRate',
         'Over18',
         'StandardHours',
         'PerformanceRating'],
        axis = 1, inplace = True)
```

```python
df.info()
```

```python
order_col = ['BusinessTravel', 'JobRole', 'EducationField']
```

```python
nominal_col = ['OverTime', 'Department', 'Gender', 'MaritalStatus']
```

```python
df.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order = {
    'BusinessTravel': ['Non-Travel', 'Travel_Rarely', 'Travel_Frequently'],
    'JobRole': ['Human Resources', 'Laboratory Technician', 'Healthcare Representative', 'Sales Representative', 'Sales Executive', 'Manager', 'Manufacturing Director', 'Research Scientist', 'Research Director'],
    'EducationField': ['Human Resources', 'Other', 'Technical Degree', 'Marketing', 'Medical', 'Life Sciences']
}

order_col = list(order.keys())

ordinal_encoder = OrdinalEncoder(categories= [order[col] for col in order_col])
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_enc = OneHotEncoder(drop='first', handle_unknown='ignore')
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer([
    ('ordinal', ordinal_encoder, order_col),
    ('nominal', one_hot_enc, nominal_col)], remainder= 'passthrough')
```

```python
preprocessor.fit_transform(df)
```

```python
X = df.drop('Attrition', axis= 1)  # Features
y = df['Attrition'] # Target
```

```python
x_transformed = preprocessor.fit_transform(X)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df_x = pd.DataFrame(x_transformed, columns= preprocessor.get_feature_names_out())
df_x.head()
```

```python
df_x.columns = (
    df_x.columns
    .str.strip()                          # remove spaces at start/end
    .str.lower()                          # convert to lowercase
    .str.replace(r"\(.*\)", "", regex=True)  # remove any text inside brackets
    .str.replace(" ", "_")                # replace spaces with underscores
    .str.replace("__", "_")               # collapse double underscores
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train_transformed, y_train)
```

```python
y_pred = linear_model.predict(X_test_transformed)
y_pred
```

```python
linear_model.score(X_test_transformed, y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error,r2_score,mean_absolute_percentage_error

y_hat = linear_model.predict(X_test_transformed)

mse = mean_squared_error (y_test, y_hat)
mae = mean_absolute_error(y_test, y_hat)
rmse = root_mean_squared_error(y_test, y_hat)
r2 = r2_score (y_test, y_hat)
mape = mean_absolute_percentage_error (y_test, y_hat)

print(f"mse : {mse}")
print(f"mae : {mae}")
print(f"rmse : {rmse}")
print(f"r2 : {r2}")
print(f"mape : {mape}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_hat1)
cm
```
